```python
import matplotlib.pyplot as plt
from skimage import io, color, exposure
```

```python
image = io.imread('./asset/eminem.webp')
# Konversi gambar ke grayscale jika gambar berwarna
if len(image.shape) == 3:
 image = color.rgb2gray(image)
# Menghitung histogram
hist, hist_centers = exposure.histogram(image)
# Plot histogram
plt.figure(figsize=(8, 6))
plt.plot(hist_centers, hist, lw=2)
plt.title("Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Berdasarkan Hasil dari Analasisi Kami gambar cenderung lebih condong ke Cerah di banding Gelap


```python
image = io.imread('./asset/eminem.webp')
# Konversi gambar ke grayscale jika gambar berwarna
if len(image.shape) == 3:
 image = color.rgb2gray(image)
# Melakukan histogram equalization
image_eq = exposure.equalize_hist(image)
# Plot gambar asli dan hasil histogram equalization
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Gambar Asli')
ax[0].axis('off')
ax[1].imshow(image_eq, cmap='gray')
ax[1].set_title('Gambar Setelah Histogram Equalization')
ax[1].axis('off')
plt.show()
```
